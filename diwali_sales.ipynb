```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # it is used for visualizing data.
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape') # to avoid encoding error use 'unicode_escape' .
```

```python
df.shape # df means dataframe
```

```python
df.head(10)
```

```python
# now imp step data cleaning.
df.info()
```

```python
# drop unused or unrelated / blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
pd.isnull(df)
# it will give values in true and false
```

```python
# check for null values
pd.isnull(df).sum()
# it will give total sum of null values
# in output u will see all values 0 but 12 in amount.
```

```python
# in amount we have value 12. we have to delete this how we will do.
# we will use
# drop null values
df.dropna(inplace=True)
```

```python
df.shape
# here rows are deleted u can see
```

```python
pd.isnull(df).sum()
# here u can check all values become 0.
```

```python
# what is the use of (inplace=True)?
# here we will see by taking an example.
# initialize list of lists
data_test = [['madhav', 11], ['Gopi', 15], ['keshav', ], ['Lalita', 16]]

# create the pandas dataframe using list
df_test = pd.DataFrame(data_test, columns=['Name', 'Age'])

df_test
```

```python
# here we didn't assign value to dropna so it will only remove na values from that particular site and not from whole data
df_test.dropna
# here keshav column is droped
# (inplace=True) it is used to save the data after drop
# we can create any variable to store it
```

```python
df_test.dropna(inplace=True)
```

```python
df_test
# here u can see keshav colummn is deleted from data
```

```python
# now we will change data type
df['Amount'] = df['Amount'].astype('int')
# here amount is float means decimal values and we have to change it to whole values.
```

```python
# here u can see data type is changed.
df['Amount'].dtypes
```

```python
# now to check columns in dataframe
# we use 
df.columns
# here it will show all columns in dataframe
```

```python
# now we will rename column
df.rename(columns= {'Marital_Status':'Shaadi'})
# here u can see column name is changed to shaadi from marital Status
# here we have not used (inplace=True) so it will not be saved and will changed to default.
```

```python
# now describe() method returns description of the data in the dataframe( i.e. count, mean, std, etc) 
df.describe()
```

```python
# use describe() for specific columns
df[['Age', 'Orders', 'Amount']].describe()
# here u can see in output only on 3 columns describe worked
```

#  Here we will start exploratory Data Analysis

# gender

```python
sns.countplot(x = 'Gender',data = df)
```

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers: # we have made a container so that to store values
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y= 'Amount', data = sales_gen)
```

# from above graphs we can see that most of the buyers are females and purchasing power of females are more than men

```python
# total amount vs total age group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group' ,y= 'Amount' ,data = sales_age)
```

# State

```python
# total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State', y= 'Orders')
```

```python
# here we do it according to how much money spend on orders or we can say (Amount).
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State', y= 'Amount')
```

# Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers: # we have made a container so that to store values
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data = sales_state, x = 'Marital_Status', y= 'Amount', hue='Gender')
```

# Occupation

```python
ax = sns.countplot(data = df, x = 'Occupation')

sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers: # we have made a container so that to store values
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data = sales_state, x = 'Occupation', y= 'Amount')
```

# here we can see most of the buyers are working in IT, Aviation and healthcare

Product category

```python
ax = sns.countplot(data = df, x = 'Product_Category')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers: # we have made a container so that to store values
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data = sales_state, x = 'Product_Category', y= 'Amount')
```

# here we can see most of the products are from food, footwear and electronics
